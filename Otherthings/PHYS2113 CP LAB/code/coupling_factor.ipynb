```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
def K_outphase(omega_n, omega_p):
    K_o = omega_p**2/(omega_n**2 + omega_p**2)
    return K_o
```

```python
def K_beat(omega_min, omega_pls):
    K_b = 4*(omega_pls**2-omega_min**2)/((omega_min+omega_pls)**2-2*(omega_min+omega_pls))
    return K_b
```

```python
#Theoratical coupling constants
#Out of phase
L1 = np.array([0.6, 0.4, 0.2])
L2 = np.array([0.6, 0.4])
K_op_t = np.array([K_outphase(3.18, 3.38), K_outphase(3.18,3.27), K_outphase(3.18, 3.20)])
#beat mode
K_bm_t = np.array([K_beat(3.18,3.36), K_beat(3.18, 3.27), K_beat(3.18, 3.20)])

#Experimental coupling constant
#Out of phase
K_op_e = np.array([K_outphase(2.85, 3.14), K_outphase(2.85,3.02), K_outphase(2.85, 3.09)])
#Beat mode
K_bm_e = np.array([K_beat(2.86,3.16), K_beat(2.87, 3.01)])
```

```python
res_op_t = stats.linregress(L1, K_op_t)
res_bm_t = stats.linregress(L1, K_bm_t)
res_op_e = stats.linregress(L1, K_op_e)
res_bm_e = stats.linregress(L2, K_bm_e)
```

```python
plt.plot(L1, res_op_t.intercept + res_op_t.slope*L1, 'r', label='Theoratical Out of Phase')
plt.plot(L1, res_bm_t.intercept + res_bm_t.slope*L1, 'b', label='Theoratical Beat Mode')
plt.plot(L1, res_op_e.intercept + res_op_e.slope*L1, 'tab:purple', label='Experimental Out of Phase')
plt.plot(L2, res_bm_e.intercept + res_bm_e.slope*L2, 'k', label='Experimental Beat Mode')
plt.title("Coupling Factor")
plt.legend(loc = 'best')
plt.xlabel('Coupling Position (m)')
plt.ylabel('Coupling Factor')
plt.plot()
```

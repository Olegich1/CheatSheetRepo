```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
def K_outphase(omega_n, omega_o):
    K_out = omega_o**2/(omega_n**2 + omega_o**2)
    return K_out
```

```python
def K_beat(omega_n, omega_o):
    K_b = omega_o**2/(omega_n**2 + omega_o**2)
    return K_b
```

```python
#Theoratical coupling constants
L1 = np.array([0.6, 0.4, 0.2])
K_t = np.array( [K_outphase(3.18, 3.35), K_outphase(3.18,3.25), K_outphase(3.18, 3.20)] )

#Experimental coupling constant
a = (K_outphase(2.85, 3.14)+K_beat(2.86,3.16))/2
b = (K_outphase(2.85,3.02) + K_beat(2.87, 3.01))/2
K_e = np.array([a, b, K_outphase(2.85, 3.09)])
```

```python
res_t = stats.linregress(L1, K_t)
res_e = stats.linregress(L1, K_e)

print('Theoratical Slope: ' + str(res_t.slope) + ' +/- ' +str(res_t.stderr))

print('Experimental Slope: ' + str(res_e.slope) + ' +/- ' +str(res_e.stderr))
```

```python
plt.plot(L1, res_t.intercept + res_t.slope*L1, 'r', label='Theoratical K slope')
plt.plot(L1, res_e.intercept + res_e.slope*L1, 'b', label='Experimental K slope')
plt.plot(L1, K_t, 'r*', label = "Theoratical K")
plt.plot(L1, K_e, 'bs', label = "Experimental K" )
plt.title('Coupling Position v.s. Coupling Factor')

plt.legend(loc = 'best')
plt.xlabel('Coupling Position (m)')
plt.ylabel('Coupling Factor')
plt.plot()
```

# Some Calculations
## My notaions
        Just some notaions here.
$$
\begin{aligned}
    &\text{Characteristic Frequency of a pendulum: } &{\omega_p} &= \sqrt{\frac{g}{L}};\\
    &\text{Coupling Constant: } &{\omega_s} &= \sqrt{\frac{kl^2}{mL^2}};
\end{aligned}
$$





## In Phase Mode

        The charateristic frequency is

$$
    \omega_i = \omega_p = \sqrt{\frac{g}{L}},
$$

    and the uncertain of that would be 
    
$$
    \Delta \omega_i = \sqrt{\left(\Delta g \frac{\partial \omega_i}{\partial g}\right)^2 + \left(\Delta L \frac{\partial \omega_i}{\partial L}\right)^2}.
$$

        Simplifying that, it is just
    
$$
    \Delta \omega_i = \sqrt{\left(\frac{\Delta g}{2\sqrt{gL}}\right)^2 + \left(\frac{\Delta L \sqrt{g}}{2L^{3/2}}\right)^2}.
$$

## Opposite Phase Mode
        The charateristic frequency is

$$
    \omega_o = \sqrt{{\omega_p}^2+2{\omega_s}^2} = \sqrt{\frac{g}{L} + \frac{kl^2}{mL^2}}.
$$

        The uncertainty is

$$
    \Delta \omega_o = \sqrt{\left(\Delta g \frac{\partial \omega_o}{\partial g}\right)^2 + \left(\Delta L \frac{\partial \omega_o}{\partial L}\right)^2 + \left(\Delta l \frac{\partial \omega_o}{\partial l}\right)^2 +\left(\Delta k \frac{\partial \omega_o}{\partial k}\right)^2 + \left(\Delta m \frac{\partial \omega_o}{\partial m}\right)^2}.
$$

        Computing this (painfully), obtaining 

$$
\Delta \omega_o=\sqrt{\frac{\text{$\Delta $l}^2 k^2 l^2}{L^4 m^2 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $m}^2 k^2 l^4}{4 L^4 m^4 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $k}^2 l^4}{4 L^4 m^2 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $g}^2}{4 L^2 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $L}^2 \left(-\frac{g}{L^2}-\frac{2 k l^2}{L^3 m}\right)^2}{4 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}}
$$

## Beat Mode (Right beat Left) for weak coupling
        The charateristic frequencies are (two terms)

$$
\begin{aligned}
    \omega_- &= \omega_p\\
    \omega_+ &= \sqrt{{\omega_p}^2+2{\omega_s}^2} = \sqrt{\frac{g}{L} + \frac{kl^2}{mL^2}}.
\end{aligned}
$$

        The uncertainties for both terms are
$$
\begin{aligned}
    \Delta \omega_- &= \sqrt{\left(\Delta g \frac{\partial \omega_-}{\partial g}\right)^2 + \left(\Delta L \frac{\partial \omega_-}{\partial L}\right)^2}.\\
    \Delta \omega_+ &= \sqrt{\left(\Delta g \frac{\partial \omega_+}{\partial g}\right)^2 + \left(\Delta L \frac{\partial \omega_+}{\partial L}\right)^2 + \left(\Delta l \frac{\partial \omega_+}{\partial l}\right)^2 +\left(\Delta k \frac{\partial \omega_+}{\partial k}\right)^2 + \left(\Delta m \frac{\partial \omega_+}{\partial m}\right)^2}.
\end{aligned}
$$

        Computing those (even more painfully), obtaining

$$
\begin{aligned}
    \Delta \omega_- &= \sqrt{\left(\frac{\text{$\Delta $L} \sqrt{g}}{2 L^{3/2}}\right)^2+\left(\frac{\text{$\Delta $g}}{2 \sqrt{g L}}\right)^2}\\
    \Delta \omega_+ &= \sqrt{\frac{\text{$\Delta $l}^2 k^2 l^2}{L^4 m^2 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $m}^2 k^2 l^4}{4 L^4 m^4 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $k}^2 l^4}{4 L^4 m^2 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $g}^2}{4 L^2 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}+\frac{\text{$\Delta $L}^2 \left(-\frac{g}{L^2}-\frac{2 k l^2}{L^3 m}\right)^2}{4 \left(\frac{g}{L}+\frac{k l^2}{L^2 m}\right)}}. 
\end{aligned}
$$

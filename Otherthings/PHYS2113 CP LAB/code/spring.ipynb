```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
force = np.array([0E-3,10E-3,20E-3,30E-3,40E-3,50E-3,60E-3])*9.8
length = np.array([10E-2,15.5E-2,18.0E-2,21.5E-2,24.5E-2,28.0E-2,31.0E-2])
```

```python
res = stats.linregress(length, force)
plt.plot(length, force/9.8, 'o', label = 'original data')
plt.plot(length, res.intercept + res.slope*length, 'r', label='fitted line')
plt.legend()
plt.title("Spring Constant")
plt.xlabel('Mass (kg)')
plt.ylabel('Length (m)')
plt.show()
print('Slope: ' + str(res.slope) + '\n' + "Sigma: " + str(res.stderr/2))
```
